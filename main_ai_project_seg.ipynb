```python
import os

# Path to the file we need to fix
file_path = "/usr/local/envs/mevis_project/lib/python3.8/site-packages/torch/utils/cpp_extension.py"

try:
    with open(file_path, 'r') as f:
        lines = f.readlines()

    new_lines = []
    found = False

    for line in lines:
        # Find the line that starts the error
        if "raise RuntimeError(CUDA_MISMATCH_MESSAGE.format" in line:
            # Get the indentation from the original line
            indentation = line[:len(line) - len(line.lstrip())]
            # Replace it with a pass statement, keeping the same indentation
            new_lines.append(f"{indentation}pass # Patched to bypass CUDA mismatch\n")
            found = True
        else:
            new_lines.append(line)

    # Write the file back out
    if found:
        with open(file_path, 'w') as f:
            f.writelines(new_lines)
        print("Successfully patched cpp_extension.py!")
    else:
        print("ERROR: Could not find the line to patch.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# 1. Install libraries needed for translation
!pip install sentencepiece transformers

# 2. Import the translator
from transformers import pipeline
import string

# 3. Load the multi-language translation model
# This single model can translate over 200 languages to English ("en")
print("Loading multi-language translation model...")
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-mul-en")
print("Translator loaded!")


# 4. Define your "Text Accuracy Enhancement" function
def enhance_prompt(raw_text):
    """
    Cleans a raw text prompt to make it ideal for the model.
    1. Converts to lowercase.
    2. Removes punctuation.
    3. Strips extra whitespace.
    """
    # Convert to lowercase
    clean_text = raw_text.lower()

    # Remove punctuation (e.g., "!!!", "?")
    clean_text = clean_text.translate(str.maketrans('', '', string.punctuation))

    # Remove leading/trailing whitespace
    clean_text = clean_text.strip()

    # Optional: You could add more rules here, like removing stop words
    # (but for a model like BERT, this is usually not needed)

    return clean_text

print("Enhancement functions are ready.")
```

```python
# --- STEP A: YOUR ENHANCEMENT MODULE ---

# 1. Get the user's raw input (in any language)
# Try changing this to Spanish ("un hipopótamo"), German ("ein Nilpferd"), etc.
user_input = "car on the right side"

print(f"Original User Input: '{user_input}'")

# 2. Translate it to English
# We re-use the 'translator' variable we created in the cell above
translated_text = translator(user_input)[0]['translation_text']
print(f"Translated to English: '{translated_text}'")

# 3. Enhance the translated text for accuracy
final_prompt = enhance_prompt(translated_text)
print(f"Final, Enhanced Prompt: '{final_prompt}'")


# --- STEP B: RUN THE MODEL WITH YOUR ENHANCED PROMPT ---

# We pass the 'final_prompt' variable into the command
!python grounded_sam2_tracking_demo_custom_video_input_gd1.0_hf_model.py \
  --config Grounded-SAM-2/grounded_sam2_cfg/grounded_sam2_swin-t_cfg.py \
  --gd_ckpt /content/Grounded-SAM-2/gdino_checkpoints/groundingdino_swint_ogc.pth \
  --sam_ckpt /content/GGrounded-SAM-2/checkpoints/sam2_hiera_tiny.pt \
  --input_video "./assets/tracking_car.mp4" \
  --text_prompt "{final_prompt}" \
  --output_dir "outputs_enhanced_test" \
  --device "cuda"
```

```python
# THIS IS THE FIX:
# Manually add the cloned 'samurai' repo to the Python path
import sys
sys.path.append('/content/samurai')

import torch
import cv2
import supervision as sv
from groundingdino.util.inference import Model as GroundingDINO
from sam2.build_sam import build_sam2
from samurai.samurai import SAMURAI  # This will now work

# --- 1. SETTINGS ---
DEVICE = torch.device("cuda")
VIDEO_PATH = "/content/samurai/assets/breakdance.mp4"
TEXT_PROMPT = "a person"  # You can change this!
GROUNDING_DINO_CHECKPOINT_PATH = "/content/weights/groundingdino_swint_ogc.pth"
SAM2_CHECKPOINT_PATH = "/content/weights/sam2_hiera_large.pt"
OUTPUT_VIDEO_PATH = "/content/output.mp4"

# --- 2. LOAD MODELS ---
print("Loading models...")
# Load GroundingDINO (Language Encoder)
grounding_dino = GroundingDINO(
    model_config_path="groundingdino/config/GroundingDINO_SwinT_OGC.py",
    model_checkpoint_path=GROUNDING_DINO_CHECKPOINT_PATH,
    device=DEVICE
)
# Load SAM 2 (Visual Backbone)
sam2_model = build_sam2(
    model_id="sam2_hiera_l",
    ckpt_path=SAM2_CHECKPOINT_PATH,
).to(DEVICE)

# --- 3. GET FIRST FRAME & DETECT OBJECT ---
print(f"Finding '{TEXT_PROMPT}' in the first frame...")
video_info = sv.VideoInfo.from_video_path(VIDEO_PATH)
frame_generator = sv.get_video_frames_generator(VIDEO_PATH)
frame = next(frame_generator)

# Use GroundingDINO to find the object from text
detections = grounding_dino.predict_with_classes(
    image=frame,
    classes=[TEXT_PROMPT],
    box_threshold=0.3,
    text_threshold=0.3
)
# Get the first (best) detection
if len(detections.xyxy) == 0:
    print(f"Error: Could not find '{TEXT_PROMPT}' in the first frame.")
else:
    first_box = detections.xyxy[0]
    print(f"Object found at: {first_box}")

    # --- 4. INIT TRACKER & PROCESS VIDEO ---
    print("Initializing SAMURAI tracker...")
    tracker = SAMURAI(sam2_model)
    tracker.register_template(frame, first_box)

    # Prepare video writer
    with sv.VideoSink(OUTPUT_VIDEO_PATH, video_info) as sink:
        # Write the first frame
        annotated_frame = sv.draw_xyxy(frame, first_box)
        sink.write_frame(annotated_frame)

        # Loop through the rest of the video
        print("Tracking object through video (this will take a few minutes)...")
        for i, frame in enumerate(frame_generator):
            if i % 50 == 0:
                print(f"  Processing frame {i}...")

            # Track the object
            pred_box, _, _ = tracker.track(frame)

            # Draw the box and save the frame
            annotated_frame = sv.draw_xyxy(frame, pred_box)
            sink.write_frame(annotated_frame)

    print("---")
    print(f"✅✅✅ Success! Video saved to {OUTPUT_VIDEO_PATH} ✅✅✅")
    print("You can download the 'output.mp4' file from the Colab explorer on the left.")
```
